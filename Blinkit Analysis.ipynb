### **BUSINESS REQUIREMENTS**

### **KPI's REQUIREMENTS**

```python
#Total Sales
total_sales = df['Sales'].sum()

#Averae Sales
avg_sales = df['Sales'].mean()

#No of Items Sold 
no_of_items_sold = df['Sales'].count()

#Average Ratings
avg_ratings = df['Rating'].mean()

#Display 

print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.0f}")
print(f"No of Items Sold: {no_of_items_sold:,.0f}")
print(f"Average Ratings: {avg_ratings:,.1f}")
```

### **CHARTS REQUIREMENTS**

#### **Total Sales by Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat, labels = sales_by_fat.index,
                       autopct = '%.1f%%',
                       startangle = 90)
plt.title('Sales by Fat Content')
plt.axis('equal')
plt.show()
```

### **Total Sales by Item Type**

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
                           
plt.figure(figsize=(10, 6))

bars = plt.bar(sales_by_type.index, sales_by_type.values)

plt.xticks(rotation =-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}',ha='center', va='bottom',fontsize=8)

plt.tight_layout()
plt.show()
```

### **Fat Content by Outlet for Total Sales**

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]

ax = grouped.plot(kind='bar', figsize=(8,5), title='Oultlet Tier by Item Fat Content')
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

 ### **Total Sales by Outlet Establishment**

```python
sales_by_year = (df.groupby('Outlet Establishment Year').sum().sort_index())

plt.figure(figsize=(9,5))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')

for x,y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x,y,f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

### **Sales by Outlet Size**

### **Sales by Outlet Location**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales',ascending = False)

plt.figure(figsize=(8,3)) #small height, enough width
ax = sns.barplot(x='Sales',y='Outlet Location Type', data=sales_by_location)

plt.title('Total Sales by Outslet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout() #Ensures layout fits without scroll
plt.show()
```

### **Product Diversity by Outlet Size**

### **Pareto (8020) Sales Contribution**

```python
print(list(df.columns))
```

```python
import matplotlib.pyplot as plt

# Clean column names
df.columns = df.columns.str.strip()

# Group by item and calculate total sales
item_sales = (
    df.groupby('Item_Identifier')['Sales']
      .sum()
      .sort_values(ascending=False)
)

# Cumulative sales percentage
cum_sales = item_sales.cumsum() / item_sales.sum()

# Plot Pareto chart
plt.figure(figsize=(7,4))
plt.plot(cum_sales.values)
plt.axhline(0.8, linestyle='--', color='red')
plt.xlabel('Items')
plt.ylabel('Cumulative Sales %')
plt.title('Pareto Analysis (8020 Rule)')
plt.tight_layout()
plt.show()
```

### **Sales Volatility by Item Type**

```python
print(df.columns.tolist())
df.columns = df.columns.str.strip()
```

```python
import matplotlib.pyplot as plt

# Clean column names
df.columns = df.columns.str.strip().str.replace(' ', '_')

# Calculate sales volatility (standard deviation of sales)
volatility = (
    df.groupby('Item_Type')['Item_Weight']
      .std()
      .sort_values(ascending=False)
      .head(10)
)

# Plot
plt.figure(figsize=(9, 4))
plt.bar(volatility.index, volatility.values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Item Type')
plt.ylabel('Sales Volatility (Std Dev)')
plt.title('Top 10 Item Types by Sales Volatility')
plt.tight_layout()
plt.show()
```

### **Outlet Efficiency Analysis**

```python
efficiency = (
    df.groupby('Outlet_Type')['Item_Visibility']
      .mean()
      .sort_values()
)

plt.figure(figsize=(8,4))
plt.barh(efficiency.index, efficiency.values)
plt.xlabel('Average Sales per Item')
plt.title('Outlet Efficiency Analysis')
plt.tight_layout()
plt.show()
```

### **Sales Variability**

```python
sales_std = (
    df.groupby('Outlet_Size')['Item_Visibility']
    .std()
    .sort_values(ascending=False)
    .head(8)
)

plt.figure(figsize=(8,4))
plt.bar(sales_std.index, sales_std.values)
plt.xticks(rotation=45)
plt.title('Sales Variability by Item Type')
plt.ylabel('Sales Std Deviation')
plt.tight_layout()
plt.show()
```

### **Category Performance by Location**

```python
tier3 = df[df['Outlet_Location_Type'] == 'Tier 3']

top_cat = (
    tier3.groupby('Item_Type')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(8)
)

plt.figure(figsize=(8,4))
plt.bar(top_cat.index, top_cat.values)
plt.xticks(rotation=45)
plt.title('Top Categories in Tier 3 Locations')
plt.tight_layout()
plt.show()
```

### **Sales Efficiency Matrix**

```python
import matplotlib.pyplot as plt

avg_visibility = df["Item_Visibility"].mean()
avg_sales = df["Sales"].mean()

plt.figure()
plt.scatter(df["Item_Visibility"], df["Sales"])
plt.axvline(avg_visibility)
plt.axhline(avg_sales)

plt.text(0.02, avg_sales+5, "Low Visibility\nHigh Sales", fontsize=9)
plt.text(avg_visibility+0.01, avg_sales+5, "High Visibility\nHigh Sales", fontsize=9)
plt.text(0.02, 20, "Low Visibility\nLow Sales", fontsize=9)
plt.text(avg_visibility+0.01, 20, "High Visibility\nLow Sales", fontsize=9)

plt.title("Sales Efficiency Matrix")
plt.xlabel("Item_Visibility")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

### **Outlet Maturity Impact Curve**

```python
df["Outlet Age"] = 2025 - df["Outlet_Establishment_Year"]

age_sales = df.groupby("Outlet Age")["Sales"].mean()

plt.figure()
age_sales.plot()
plt.title("Outlet Maturity Impact on Sales")
plt.xlabel("Outlet Age (Years)")
plt.ylabel("Average Sales")
plt.tight_layout()
plt.show()
```

### **Product RiskReward Chart**

```python
risk_reward = df.groupby("Item_Type").agg(
    Avg_Sales=("Sales", "mean"),
    Sales_Volatility=("Sales", "std")
)

plt.figure()
plt.scatter(risk_reward["Sales_Volatility"], risk_reward["Avg_Sales"])

for i in risk_reward.index:
    plt.text(risk_reward.loc[i, "Sales_Volatility"],
             risk_reward.loc[i, "Avg_Sales"],
             i, fontsize=8)

plt.title("Product RiskReward Analysis")
plt.xlabel("Sales Volatility (Risk)")
plt.ylabel("Average Sales (Reward)")
plt.tight_layout()
plt.show(
)
```

### **Hidden Champions Chart**

```python
hidden = df[
    (df["Item_Visibility"] < df["Item_Visibility"].median()) &
    (df["Sales"] > df["Sales"].quantile(0.75))
]

top_hidden = hidden.groupby("Item_Type")["Sales"].mean().sort_values(ascending=False)

plt.figure()
top_hidden.plot(kind="bar")
plt.title("Hidden Champion Products")
plt.xlabel("Item Type")
plt.ylabel("Average Sales")
plt.tight_layout()
plt.show()
```

### **Outlet Performance Score**

```python
outlet_score = df.groupby("Outlet_Identifier").agg(
    Avg_Sales=("Sales", "mean"),
    Avg_Rating=("Rating", "mean"),
    Avg_Visibility=("Item_Visibility", "mean")
)

outlet_score["Performance Score"] = (
    outlet_score["Avg_Sales"] * 0.5 +
    outlet_score["Avg_Rating"] * 30 -
    outlet_score["Avg_Visibility"] * 100
)

top_outlets = outlet_score.sort_values("Performance Score", ascending=False).head(10)

plt.figure()
top_outlets["Performance Score"].plot(kind="bar")
plt.title("Top 10 Outlet Performance Score")
plt.xlabel("Outlet")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```
